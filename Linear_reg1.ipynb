```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/singhrau0/Big-Data-Preprocessing/main/Movie_Data.csv')
```

```python
df
```

```python
# finding correlation matrix
df.corr()
```

```python
# for coluring use heatmap
sns.heatmap(df.corr())
# We have white for budget and trailer views
```

```python
# Independent variable
x=df[['Trailer_views','Budget']]
```

```python
# Dependent variable
y=df['Collection']
```

```python
# dividing dataset using methods.Train and test data.
```

```python
from sklearn.model_selection import train_test_split
```

```python
# To control shuffling fourth argument
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2529)
# as we have to compare the values train test we are fixing the shuffle
```

```python
x_train
```

```python
x_test
```

```python
# using linear regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
# Fitting the data in model
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
x_test.head(1)
```

```python
# for predicting
lr.predict(x_test.head(1))
```

```python
# For range iloc for specific value loc
x_test.loc[[461]]
```

```python
# for intercept
lr.intercept_
```

```python
lr.predict(x_test.loc[[461]])
```

```python
# For getting accuracy score
lr.score(x_test,y_test)
```

```python
# for slope
lr.coef_
```
